```python
text = """Isaac Newton published Philosophiæ Naturalis Principia Mathematica in 1687 in London, laying the foundation of classical mechanics.
Mount Everest stands at 8,848.86 meters and lies on the border between Nepal and China.
Marie Curie won 2 Nobel Prizes (1903 and 1911), becoming the first person to achieve this distinction.
Tokyo has a population of over 37 million people, making it one of the largest metropolitan areas in the world.
Apollo 11 Moon Landing occurred on July 20, 1969, when Neil Armstrong became the first human to walk on the Moon.
Apple Inc. was founded in 1976 by Steve Jobs, Steve Wozniak, and Ronald Wayne in Cupertino.
Sahara Desert covers about 9.2 million square kilometers across Africa.
Albert Einstein was awarded the Nobel Prize in 1921 for his explanation of the Photoelectric Effect.
Amazon River is approximately 6,400 km long and flows through countries like Brazil and Peru.
Google was founded in 1998 by Larry Page and Sergey Brin while they were PhD students at Stanford University"""
type(text)
```

```python
import spacy
```

```python
#load the language model
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp(text)
```

```python
#stemming
import nltk
from nltk.stem.porter import *

p_stem=PorterStemmer()
for word in text.split():
  print(word,"-->",p_stem.stem(word))
```

```python
#lemmatization
for token in nlp(text):
  print(token.text,"-> ",token.pos_,
        "->",token.lemma_)
```
