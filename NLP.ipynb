```python
text = """niki majithiya is currently sitting in MA104 learning nlp by professor Nishith Kotak"""
type(text)
```

```python
import spacy
```

```python
#load the language model
nlp=spacy.load('en_core_web_sm')
```

```python
doc=nlp(text)
```

```python
#Named Entity Recognition: NER
for token in doc.ents:
  print(token.text, token.label_)
```

```python
#POS Tags: Part of Speech Tagging
for token in doc:
  print(token.text,"-->", token.pos_)
```

```python
#Stemming vs lemmatization
text="""I studied AI subject and
then meeting Mr. Sachin in a meeting"""
```

```python
#stemming
import nltk
from nltk.stem.porter import *

p_stem=PorterStemmer()
for word in text.split():
  print(word,"-->",p_stem.stem(word))
```

```python
#lemmatization
for token in nlp(text):
  print(token.text,"-> ",token.pos_,
        "->",token.lemma_)
```
