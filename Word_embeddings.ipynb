```python
sentences=[
     "India is the 4th largest economy in the world.",
     "India has a huge population of 1.4 billion people making it the world's most populated country surpassing China.",
     "India is highly advance in Science and Technology.",

     "Football is a good and interesting sport.",
     "I grew up watching football.",

     "India is emerging in the field of VLSI.",
     "VLSI is the manufacturing of semiconductors used in all electronic devices."
     ]

```

```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer  #TF-BoW
from sklearn.feature_extraction.text import TfidfVectorizer  #TF-IDF
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Embeddings using TF-BoW approach
tf_vectorizer = CountVectorizer()
tf_matrix = tf_vectorizer.fit_transform(sentences)
```

```python
tf_matrix
```

```python
tf_similarity=cosine_similarity(tf_matrix)
tf_similarity
```

```python
#Embeddings using TF-IDF approach
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)
```

```python
tfidf_matrix
```

```python
tfidf_similarity=cosine_similarity(tfidf_matrix)
tfidf_similarity
```

```python
#finding out the document similarity
def doc_sim(similarity_matrix,sentences):
  for i in range(len(sentences)):
    sims=similarity_matrix[i].copy()
    sims[i]=-1         #excludes the sentences that is being compared with the rest of the sentences
    top_index=np.argmax(sims)
    print(sentences[i],"--->",sentences[top_index])
```

```python
doc_sim(tfidf_similarity,sentences)
```
