```python
corpus = """
I like machine learning
I like deep learning
I enjoy machine learning
I enjoy data science
Machine learning is interesting
Deep learning is powerful
"""
```

```python
#importing basic libraries
from nltk.util import ngrams
from collections import defaultdict,Counter
import random

def get_probabilistic_next_word(current_word, prediction_model):
    current_word = current_word.lower()
    if current_word in prediction_model:
        next_word_counts = prediction_model[current_word]
        candidates = list(next_word_counts.keys())
        weights = list(next_word_counts.values())
        if candidates:
            return random.choices(candidates, weights=weights, k=1)[0]
    return None # No prediction or no candidates

def generate_text(seed_phrase, prediction_model, length=10):
    generated_text = seed_phrase.lower().split()
    current_word = generated_text[-1]

    for _ in range(length - len(generated_text)):
        next_word = get_probabilistic_next_word(current_word, prediction_model)
        if next_word:
            generated_text.append(next_word)
            current_word = next_word
        else:
            break
    return " ".join(generated_text)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
tokens=nltk.word_tokenize(corpus.lower())
```

```python
#build ngram model
n=2
bigrams=list(ngrams(tokens,n))
print(bigrams)
```

```python
#create the probability model
model=defaultdict(Counter)
for w1,w2 in bigrams:
  model[w1][w2]+=1
model
```

```python
#function to predict the next word
def predict_next_word(word):
  word=word.lower()
  if word in model:
    next_words=model[word]
    prediction=max(next_words,key=next_words.get)
    return prediction

  else:
    return "No preditcion is done"
```
